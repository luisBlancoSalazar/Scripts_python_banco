```python
import pandas as pd
```

```python
df = pd.read_excel('./sourceHelen/CredRefinanciamientosParalelosReprogramaciones_20250831.xlsx')
```

```python
# Esto es para que el ordenamiento funcione correctamente.
df['FECHA DE DESEMBOLSO / APLICACION DE REPROGRAMACION'] = pd.to_datetime(df['FECHA DE DESEMBOLSO / APLICACION DE REPROGRAMACION'], dayfirst=True)
```

```python
# Ordenamos por 'NUMERO DE SOLICITUD' y luego por 'FECHA DE DESEMBOLSO / APLICACION DE REPROGRAMACION' de forma ascendente.
df_ordenado = df.sort_values(by=['NUMERO DE SOLICITUD', 'FECHA DE DESEMBOLSO / APLICACION DE REPROGRAMACION'], ascending=True)
```

```python
#filtrar solo aquellas con estado vigente
df_vigente = df_ordenado[~df_ordenado['ESTADO DEL CREDITO'].isin(['FINALIZADO', 'CASTIGADO'])].copy()
```

```python
df_vigente
```

```python
# Agrupar y agregar la información
df_agrupado = df_vigente.groupby("NUMERO DE SOLICITUD").agg(
    # Para cada grupo, aplicamos estas operaciones
    VECES_REPETIDO=('NUMERO DE SOLICITUD', 'count'),
    NOMBRE_CLIENTE = ('NOMBRE DEL CLIENTE', 'last'),
    AGENCIA=('AGENCIA', 'first'),
    TIPO_CREDITO=('TIPO DE CREDITO', 'last'),
    SALDO_CREDITO=('MONTO DESEMBOLSADO O REPROGRAMADO', 'last'),
    ESTADO_CREDITO=('ESTADO DEL CREDITO', 'last')
    # Añade aquí otras columnas que necesites, usando 'first' o 'last'
).reset_index()
```

```python
df_agrupado.head(30)
```

```python
df_paralelo = df[df['TIPO DE CREDITO'] == 'PARALELO'].copy()
df_refinanciamiento = df[df['TIPO DE CREDITO'].isin(['REFINANCIAMIENTO POR DIFERIMIENTO','REFINANCIAMIENTO POR DECRETO SUPREMO 5241','REFINANCIAMIENTO REGULAR'])].copy()
df_reprogramacion = df[df['TIPO DE CREDITO'].isin(['REPROGRAMACION EXCEPCIONAL','REPROGRAMACIÓN DE PAGO UNICO','REPROGRAMACION EXCEPCIONAL CON CUOTAS DIFERIDAS','REPROGRAMACION POR DECRETO SUPREMO 5241','REPROGRAMACIÓN POR EXPROMISION','REPROGRAMACIÓN REGULAR','REPROGRAMACIÓN POR DIFERIMIENTO'])].copy()
```

```python
# --- Paso 1: Agrupar por Agencia y Tipo, y contar las ocurrencias ---
# Esto crea una Serie con un índice de dos niveles (Agencia, Tipo de Crédito)
conteo_agrupado = df_refinanciamiento.groupby(['AGENCIA', 'TIPO DE CREDITO']).size()
```

```python


# (Este es tu DataFrame inicial, filtrado con los tipos de crédito que te interesan)
# df_reprogramaciones = df[df['TIPO DE CREDITO'].isin([...])].copy()



# --- Paso 2: Convertir el índice 'TIPO DE CREDITO' en columnas ---
# .unstack() pivota el nivel del índice especificado, convirtiéndolo en columnas.
# fill_value=0 reemplaza los valores NaN (si una agencia no tiene un tipo) con 0.
resumen_refinanciamiento = conteo_agrupado.unstack(level='TIPO DE CREDITO', fill_value=0)


# --- Paso 3: Calcular el total por fila (por agencia) ---
# .sum(axis=1) suma horizontalmente a través de las columnas.
resumen_refinanciamiento['total'] = resumen_refinanciamiento.sum(axis=1)


# --- Paso 4: Calcular la fila de totales al final ---
# .sum(axis=0) suma verticalmente a través de las filas.
# Añadimos esta nueva fila al DataFrame con el índice 'TOTAL'.
resumen_refinanciamiento.loc['TOTAL'] = resumen_refinanciamiento.sum(axis=0)


# (Opcional) Si quieres renombrar las columnas para que sean más cortas:
resumen_refinanciamiento.columns.name = None # Limpia el nombre del índice de columnas
resumen_refinanciamiento = resumen_refinanciamiento.rename(columns={
    'REFINANCIAMIENTO REGULAR': 'REPROGR AMACION REGULAR',
    'REFINANCIAMIENTO POR DECRETO SUPREMO 5241': 'REPROGR AMACION por decreto',
    'REFINANCIAMIENTO POR DIFERIMIENTO': 'REPROGR AMACION POR DIFERIMIENTO'
})


# Mostrar el resultado final
resumen_refinanciamiento
```

```python
# --- Paso 1: Agrupar por Agencia y Tipo, y contar las ocurrencias ---
conteo_agrupado = df_reprogramacion.groupby(['AGENCIA', 'TIPO DE CREDITO']).size()

# --- Paso 2: Convertir el índice 'TIPO DE CREDITO' en columnas ---
resumen_reprogramacion = conteo_agrupado.unstack(level='TIPO DE CREDITO', fill_value=0)

# --- Paso 3: Calcular el total por fila (por agencia) ---
resumen_reprogramacion['TOTAL'] = resumen_reprogramacion.sum(axis=1)

# --- Paso 4: Calcular la fila de totales al final ---
resumen_reprogramacion.loc['TOTAL'] = resumen_reprogramacion.sum(axis=0)

# Mostrar el resultado final
resumen_reprogramacion
```

```python
resumen_por_agencia_paralelo = df_paralelo.groupby('AGENCIA').agg(
    # Contamos cuántas solicitudes únicas tiene cada agencia
    NUMERO_DE_PARALELOS=('NUMERO DE SOLICITUD', 'count'),
    
    # Sumamos el monto desembolsado o reprogramado para obtener el total por agencia
    MONTO_DESEMBOLSADO_O_REPROGRAMADO_TOTAL=('MONTO DESEMBOLSADO O REPROGRAMADO', 'sum')
).reset_index()
```

```python
resumen_por_agencia_reprogramacion = resumen_refinanciamiento.groupby('AGENCIA').agg(
    # Contamos cuántas solicitudes únicas tiene cada agencia
    NUMERO_DE_REPROGRAMACIONES=('NUMERO DE SOLICITUD', 'count'),

    # Sumamos el monto desembolsado o reprogramado para obtener el total por agencia
    MONTO_DESEMBOLSADO_O_REPROGRAMADO_TOTAL=('MONTO DESEMBOLSADO O REPROGRAMADO', 'sum')
).reset_index()
```

```python
# 3. Eliminar duplicados
# Nos quedamos con el último registro de cada 'NUMERO DE SOLICITUD', que será el más reciente.
df_sin_duplicados = df_vigente.drop_duplicates(subset='NUMERO DE SOLICITUD', keep='last')
```

```python
# 
#df_sin_duplicados.to_excel("resumen_agencias.xlsx", index=False, sheet_name="Resumen por Agencia")
# Usar ExcelWriter para manejar el archivo
with pd.ExcelWriter('reporte_CredRefinanciamientosParalelosReprogramaciones.xlsx') as writer:
    # Escribir el primer DataFrame en una hoja
    df_vigente.to_excel(writer, sheet_name='Detalle Total', index=False)

    df_sin_duplicados.to_excel(writer, sheet_name='Detalle Unicos', index=False)
    
     # Escribir el segundo DataFrame en otra hoja
    df_agrupado.to_excel(writer, sheet_name="Resumen por NroSolicitud", index=False)
    # Escribir el tercer DataFrame en otra hoja
    resumen_refinanciamiento.to_excel(writer, sheet_name='Resumen por Refinanciamiento', index=True)

    resumen_reprogramacion.to_excel(writer, sheet_name='Resumen por Reprogramacion', index=True)

    resumen_por_agencia_paralelo.to_excel(writer, sheet_name='Resumen  Paralelo', index=False)   
```
