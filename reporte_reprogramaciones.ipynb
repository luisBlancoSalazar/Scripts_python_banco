```python
import pandas as pd
```

```python
df = pd.read_excel('./sourceHelen/Restructuraciones_20250831.xlsx')
```

```python
# Esto es para que el ordenamiento funcione correctamente.
df['FECHA DE RESTRUCTURACION'] = pd.to_datetime(df['FECHA DE RESTRUCTURACION'], dayfirst=True)
```

```python
# Ordenamos por 'NUMERO DE SOLICITUD' y luego por 'FECHA DE RESTRUCTURACION' de forma ascendente.
df_ordenado = df.sort_values(by=['NUMERO DE SOLICITUD', 'FECHA DE RESTRUCTURACION'], ascending=True)
```

```python
#filtrar solo aquellas con estado vigente
df_vigente = df_ordenado[df_ordenado['ESTADO DE CREDITO ACTUAL'] == 'VIGENTE'].copy()
```

```python
# Agrupar y agregar la información
df_agrupado = df_vigente.groupby("NUMERO DE SOLICITUD").agg(
    # Para cada grupo, aplicamos estas operaciones
    VECES_REPETIDO=('NUMERO DE SOLICITUD', 'count'),
    AGENCIA=('AGENCIA', 'first'),
    SALDO_CREDITO=('SALDO CREDITO A LA FECHA', 'last'),
    ESTADO_REESTRUCTURACION=('ESTADO DEL CREDITO A LA FECHA DE RESTRUCTURACION', 'first')
    # Añade aquí otras columnas que necesites, usando 'first' o 'last'
).reset_index()
```

```python
df_agrupado.head(30)
```

```python
resumen_por_agencia = df_agrupado.groupby('AGENCIA').agg(
    # Contamos cuántas solicitudes únicas tiene cada agencia
    NUMERO_DE_REESTRUCTURACIONES=('NUMERO DE SOLICITUD', 'count'),
      
    # Sumamos el saldo de crédito para obtener el total por agencia
    SALDO_CREDITO_TOTAL=('SALDO_CREDITO', 'sum')
).reset_index()
```

Eliminar lo que son las las columnas repetidas

```python
# 3. Eliminar duplicados
# Nos quedamos con el último registro de cada 'NUMERO DE SOLICITUD', que será el más reciente.
df_sin_duplicados = df_vigente.drop_duplicates(subset='NUMERO DE SOLICITUD', keep='last')
```

```python
# 
#df_sin_duplicados.to_excel("resumen_agencias.xlsx", index=False, sheet_name="Resumen por Agencia")
# Usar ExcelWriter para manejar el archivo
with pd.ExcelWriter('reporte_final_con_varias_hojas.xlsx') as writer:
    # Escribir el primer DataFrame en una hoja
    df_sin_duplicados.to_excel(writer, sheet_name='Detalle', index=False)
    
    # Escribir el segundo DataFrame en otra hoja
    resumen_por_agencia.to_excel(writer, sheet_name='Resumen por Agencia', index=False)
```
